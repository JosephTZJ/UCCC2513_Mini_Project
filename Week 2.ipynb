```python
def compare(x, y):
    """
    This is a function to compare 2 numbers, x and y
    """
    # if statement
    if x > y:
        print(f"{x} is greater than {y}")
    elif x < y:
        print(f"{x} is smaller than {y}")
    else:
        print(f"{x}is the same as {y}")
```

## Create custom images

```python
import matplotlib.pyplot as plt
```

```python
# create a black 2D image (grayscale image)
black_img = np.zeros((200,200), dtype = np.uint8)

plt.imshow(black_img, cmap=plt.cm.gray)
plt.title("Black Grayscale")
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# create a white image
white = 255 * np.ones((200,200))
white = np.uint8(white)

plt.imshow(white, cmap=plt.cm.gray, vmin=0, vmax=256)
plt.title("White Grayscale")
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# create a random color image
random_color_img = np.random.randint(0, high=256, size=(200, 200, 3), dtype="uint8")

plt.imshow(random_color_img)
plt.title("Random Color")
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
def OddEven(x):
    """
    This function is to determine the odd even nature of number
    """
    if isinstance(x, int):
        if x % 2 == 0:
            if x == 0:
                statement = " is neither odd or even number."
            else:
                statement = " is even number."
        else:
            statement = " is odd number."
    else:
        statement = " is not an integer!"
    
    return print(f"{x}" + statement)
```

```python
OddEven(0.9)
```

Question 4

```python
img = np.zeros((200, 200), dtype = "uint8")

# The center point of the image
center_x = 200 // 2
center_y = 200 // 2

# The radius of the circular spot
radius = 200 // 8

# Create a meshgrid of coordinates
x, y = np.meshgrid(range(200), range(200))

# range(image_size) = indices of the rows and columns in the image
# np.meshgrid() takes two 1D arrays, then returns two 2D arrays
# the code line above creates two arrays x and y, where x represents
# the x-coordinates and y represents the y-coordinates of the points on the grid.

# Calculate the  Euclidean distance from each point to the center
distance = np.sqrt((x - center_x) ** 2 + (y - center_y) ** 2)

# Set the circular spot to white
img[distance <= radius] = 1

# Display the image
plt.imshow(img, cmap= plt.cm.gray)
plt.title("Custom Image")
plt.xticks([]), plt.yticks([])
plt.show()
```
